```
punctuation=[",","।",";","?","!","—","‐",":-","'","(",")",'"']
```

```
import codecs
```

```
fp = "finalrepo/train/pmi/en-hi/train.hi"
n= 1
assert n > 0
with codecs.open(fp, encoding='utf-8') as f:
    chunk = f.readline()
    hindi.append(chunk)
```

```
from indicnlp.normalize.indic_normalize import IndicNormalizerFactory
remove_nuktas=False
#line='सब कुछ सरकार करेगी यह भाव प्रबल हो गया।'
factory=IndicNormalizerFactory()
normalizer=factory.get_normalizer("hi")
```

```
hindi1=[]
for line in hindi:
  line1=line.rstrip("\n")
  output_text=normalizer.normalize(line1)
  hindi1.append(output_text)
```

```
hindi1[::-1]
del hindi
```

```
import re
import string
from indicnlp.tokenize import indic_tokenize  
hindi2=[]

for line in hindi1:
  for t in indic_tokenize.trivial_tokenize(line):
    if(t not in punctuation and t not in stopwords):
    if(any(punct in t for punct in punctuation)==False):
      hindi2.append(t)  
          
hindi2
  

```

```
len(hindi2)
```

```
hindi3=set(hindi2)
len(hindi3)
```

```
import unicodedata as ud

latin_letters= {}

def is_latin(uchr):
    try: return latin_letters[uchr]
    except KeyError:
         return latin_letters.setdefault(uchr, 'LATIN' in ud.name(uchr))

def only_roman_chars(unistr):
    return all(is_latin(uchr)
           for uchr in unistr
           if uchr.isalpha()) # isalpha suggested by John Machin

only_roman_chars('"')


```

```
hindi3=[]
for word in hindi2:
  if(only_roman_chars(word)==False):
    hindi3.append(word)
hindi3
```

```
hindi4=set(hindi3)
hindi4
```

```
len(hindi4)
```

```
hindi5=[]
for word in hindi4:
  if(any(map(str.isdigit, word))==False):
    hindi5.append(word)
hindi5
```
