```
punctuation=[",","।",";","?","!","—","‐",":-","'","(",")",'"']
```

```
import sys
# The path to the local git repo for Indic NLP library
INDIC_NLP_LIB_HOME=r"/content/indic_nlp_library"
# The path to the local git repo for Indic NLP Resources
INDIC_NLP_RESOURCES=r"/content/indic_nlp_resources"

sys.path.append(r'{}'.format(INDIC_NLP_LIB_HOME))
from indicnlp import common
common.set_resources_path(INDIC_NLP_RESOURCES)
from indicnlp import loader
loader.load()
```

```
import codecs

from indicnlp.normalize.indic_normalize import IndicNormalizerFactory
remove_nuktas=False
fp = "finalrepo/train/pmi/en-hi/train.hi"
with open(fp) as f:
    hindi = f.readlines()
factory=IndicNormalizerFactory()
normalizer=factory.get_normalizer("hi")
```

```
import string
for char in string.punctuation:
  print(char,end=" ")
translate_table=dict((ord(char),None) for char in string.punctuation) 
```

```
import re
import string
from indicnlp.tokenize import indic_tokenize  
hindi1=[]

for line in hindi:
  line=line.rstrip("\n")
  line=normalizer.normalize(line)
  line= re.sub(r"\d+","",line)
  line=re.sub(r"[a-zA-Z]+","",line)
  line=line.translate(translate_table)
  for t in indic_tokenize.trivial_tokenize(line):
    if(t not in punctuation):
    #  if(any(punct in t for punct in punctuation)==False):
      hindi1.append(t)  
          
```

```
import collections
import pickle

from collections import Counter
counts = Counter(hindi1).most_common(10000)
with open('hindi1.pkl', 'wb') as outputfile:
        pickle.dump(counts, outputfile)


```
